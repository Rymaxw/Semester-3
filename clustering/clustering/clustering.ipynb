```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset 
iris = r"D:\Praktikum_IPSD\Dataset\Iris.csv"
df = pd.read_csv(iris)
```

```python
df.head()
```

```python
# seleksi kolom selain 'Id' dan 'Species
df_final = df.drop(columns=['Id', 'Species'])
df_final
```

```python
# membuat model K-Means dan training  data , jumlah k=3; pemilihan centroid random
from sklearn.cluster import KMeans # import Kmeans
model = KMeans(n_clusters=3, random_state=42, n_init=10).fit(df_final)
```

```python
# menampilkan label hasil clustering
df_final['Cluster'] = model.labels_
df_final.head(50)
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Cluster', data=df_final, palette='viridis', s=100)
plt.title('K-Means Clustering of Iris Dataset (Sepal Length vs Sepal Width)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
visualizer = silhouette_score(df_final.drop(columns=['Cluster']), model.labels_)
print(f'Silhouette Score: {visualizer}')
```
