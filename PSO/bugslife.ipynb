```python
import numpy as np
import matplotlib.pyplot as plt
#matplotlib.use('Jpg')
%matplotlib inline
import matplotlib.cm as cm
```

```python
n = 10000
p = 0.5
```

```python
xrandom = np.random.uniform(0,1,n)
x = [1 if k <= 0.5 else -1 for k in xrandom]
```

```python
xn=np.zeros(n)
xn[0]=0.
plt.plot(0,1, '*')
plt.ylim(0.8,1.2)
plt.xlim(-20,20)
plt.grid(axis='x', lw=1)
plt.savefig('bug{:05}.jpg'.format(0),dpi=100)
plt.show()
for i in range(1,n):
    xn[i]= xn[i-1]+x[i]
    plt.plot(xn[i],1, '*')
    plt.ylim(0.8,1.2)
    plt.xlim(-20,20)
    plt.grid(axis='x', lw=1)
    # plt.gca.axes.get_yaxis().set_visible(False)
    plt.savefig('bug{:05}.jpg'.format(i),dpi=100)
    plt.show()
```
